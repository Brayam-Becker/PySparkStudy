```python

from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df_pyspark = spark.read.csv('/content/diabetes.csv', header=True, inferSchema=True)
```

```python
spark.read.csv('/content/diabetes.csv', header=True, inferSchema=True).show()
```

```python
## drp the columns 
df_pyspark.drop('Pregnancies').show()
```

```python
df_pyspark.na.drop().show()
```

```python
## any == how
df_pyspark.na.drop(how="all").show()
```

```python
## threshold
df_pyspark.na.drop(how='all', thresh=2).show()
```

```python
df_pyspark.na.drop(how='any', subset=['Pregnancies'])
```

```python
## Filling the missing value
df_pyspark.na.fill('Missing Value', ['Pregnancies', 'Insulin']).show()
```

```python
df_pyspark.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['Pregnancies', 'Glucose', 'BloodPressure'], 
    outputCols=["{}_imputed".format(c) for c in ['Pregnancies', 'Glucose', 'BloodPressure']]
    ).setStrategy("median")
```

```python
# Add imputation cols to df
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
